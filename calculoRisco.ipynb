```python
import csv
import pandas as pd

risco = 0
###Leitura Acelerometro###
acelerometro = input("Qual o caminho para o dataset do acelerometro?")
dataAacAxisX = pd.read_csv(acelerometro, usecols=[2], header=None)
dataAacAxisY = pd.read_csv(acelerometro, usecols=[3], header=None)
dataAacAxisZ = pd.read_csv(acelerometro, usecols=[5], header=None)

###Leitura GPS###
GPS = input("Qual o caminho para o dataset do GPS?")
dataVelocidade = pd.read_csv(GPS, header=None)
velocidade = dataVelocidade[3].mean()
distanciaMetrosDoisMinutos = velocidade*120

###Leitura Magnetometro###
magnetometro = input("Qual o caminho para o dataset do magnetometro?")
dataMagAxisX = pd.read_csv(magnetometro, usecols=[2], header=None)
dataMagAxisY = pd.read_csv(magnetometro, usecols=[3], header=None)
dataMagAxisZ = pd.read_csv(magnetometro, usecols=[4], header=None)


###Leitura Giroscópio###
giroscopio = input("Qual o caminho para o dataset do giroscópio?")
dataGyrAxisX = pd.read_csv(giroscopio, usecols=[2], header=None)
dataGyrAxisY = pd.read_csv(giroscopio, usecols=[3], header=None)
dataGyrAxisZ = pd.read_csv(giroscopio, usecols=[4], header=None)

###Leitura Weather###

Weather = input("Qual o caminho para o dataset do clima?")
weatherHora = pd.read_csv(Weather)
temperatura = weatherHora.loc[0,'tempC']
visibility = weatherHora.loc[0, 'visibility']
tempMax = weatherHora.loc[0, 'maxtempC']
tempMin = weatherHora.loc[0, 'mintempC']
sensacao = weatherHora.loc[0, 'FeelsLikeC']
chuva = weatherHora.loc[0, 'precipMM']

if temperatura >=28:
  print("Está muito quente")
  risco=risco+1
elif temperatura <= 15:
  print("Está muito frio")

if visibility <=30:
  print("visibilidade boa!")
elif visibility >31 and visibility <50:
  print("visibilidade média, cuidado!")
  risco = risco + 1
elif visibility > 51:
  print("Visibilidade ruim")
  risco = risco + 2

if tempMax >= 28 or tempMin < 15:
  print("Cuidado com o clima!")
  risco = risco + 1

if sensacao >= 28:
  print("Muito quente")
  risco = risco + 1
elif sensacao <=15:
    print("Muito frio")
    risco = risco + 1

if chuva >=50:
  print("pode chover")
  risco = risco + 1



##Leitura Tráfego###

Traffic = input("Qual o caminho para o dataset do tráfego?")
trafficHora = pd.read_csv(Traffic)
duration = trafficHora.loc[1,'duration']
durationInTraffic = trafficHora.loc[1,'duration_in_traffic']
durationMed = float(duration)*1.25

if durationInTraffic <= duration:
  print("Tráfego esta bom")
  
elif durationInTraffic > duration:
  if durationInTraffic < str(durationMed):
    print("Tráfego moderado")
    risco = risco + 1
  else:
    print("Tráfego alto")
    risco = + 2


###Experiencia do ciclista###

xp = input("A quantos anos tem o costume de utilizar a bicicleta no meio urbano?")

if xp <= str(1):
  risco = risco + 2
  print("Ciclista Iniciante")
elif xp > str(1) or xp == str(2):
  risco = risco + 1
  print("Ciclista Intermediario")
elif xp > str(2):
  print("Ciclista Experiente")

###ciclovia###

ciclovia = input("O trecho tem ciclovia? sim=0 ou não=1 ")

if ciclovia == str(1):
  risco = risco +2

###Eixo X do Acelerometro a cada 2 minutos###
step = 6000

AccAxisX = dataAacAxisX.abs().groupby(dataAacAxisX.index // step).mean()
doisMinAccAxisX = AccAxisX.loc[0,2]
quatroMinAccAxisX = AccAxisX.loc[1,2]
seisMinAccAxisX = AccAxisX.loc[2,2]
oitoMinAccAxisX = AccAxisX.loc[3,2]
dezMinAccAxisX = AccAxisX.loc[4,2]

###Eixo Y do Acelerometro a cada 2 minutos###

AccAxisY = dataAacAxisY.abs().groupby(dataAacAxisY.index // step).mean()
doisMinAccAxisY = AccAxisY.loc[0,3]
quatroMinAccAxisY = AccAxisY.loc[1,3]
seisMinAccAxisY = AccAxisY.loc[2,3]
oitoMinAccAxisY = AccAxisY.loc[3,3]
dezMinAccAxisY = AccAxisY.loc[4,3]

###Eixo Z do Acelerometro a cada 2 minutos###

AccAxisZ = dataAacAxisZ.abs().groupby(dataAacAxisZ.index // step).mean()
doisMinAccAxisZ = AccAxisZ.loc[0,5]
quatroMinAccAxisZ = AccAxisZ.loc[1,5]
seisMinAccAxisZ = AccAxisZ.loc[2,5]
oitoMinAccAxisZ = AccAxisZ.loc[3,5]
dezMinAccAxisZ = AccAxisZ.loc[4,5]

doisMinAccAxisAvg = (doisMinAccAxisX + doisMinAccAxisY + doisMinAccAxisZ)/3
quatroMinAccAxisAvg = (quatroMinAccAxisX + quatroMinAccAxisY + quatroMinAccAxisZ)/3
seisMinAccAxisAvg = (seisMinAccAxisX + seisMinAccAxisY + seisMinAccAxisZ)/3
oitoMinAccAxisAvg = (oitoMinAccAxisX + oitoMinAccAxisY + oitoMinAccAxisZ)/3
dezMinAccAxisAvg = (dezMinAccAxisX + dezMinAccAxisY + dezMinAccAxisZ)/3



#########################################

###Eixo X do magnetometro a cada 2 minutos###

MagAxisX = dataMagAxisX.abs().groupby(dataMagAxisX.index // step).mean()
doisMinMagAxisX = MagAxisX.loc[0,2]
quatroMinMagAxisX = MagAxisX.loc[1,2]
seisMinMagAxisX = MagAxisX.loc[2,2]
oitoMinMagAxisX = MagAxisX.loc[3,2]
dezMinMagAxisX = MagAxisX.loc[4,2]


###Eixo Y do magnetometro a cada 2 minutos###

MagAxisY = dataMagAxisY.abs().groupby(dataMagAxisY.index // step).mean()
doisMinMagAxisY = MagAxisY.loc[0,3]
quatroMinMagAxisY = MagAxisY.loc[1,3]
seisMinMagAxisY = MagAxisY.loc[2,3]
oitoMinMagAxisY = MagAxisY.loc[3,3]
dezMinMagAxisY = MagAxisY.loc[4,3]


###Eixo Z do magnetometro a cada 2 minutos###

MagAxisZ = dataMagAxisZ.abs().groupby(dataMagAxisZ.index // step).mean()
doisMinMagAxisZ = MagAxisZ.loc[0,4]
quatroMinMagAxisZ = MagAxisZ.loc[1,4]
seisMinMagAxisZ = MagAxisZ.loc[2,4]
oitoMinMagAxisZ = MagAxisZ.loc[3,4]
dezMinMagAxisZ = MagAxisZ.loc[4,4]

doisMinMagAxisAvg = (doisMinMagAxisX + doisMinMagAxisY + doisMinMagAxisZ)/3
quatroMinMagAxisAvg = (quatroMinMagAxisX + quatroMinMagAxisY + quatroMinMagAxisZ)/3
seisMinMagAxisAvg = (seisMinMagAxisX + seisMinMagAxisY + seisMinMagAxisZ)/3
oitoMinMagAxisAvg = (oitoMinMagAxisX + oitoMinMagAxisY + oitoMinMagAxisZ)/3
dezMinMagAxisAvg = (dezMinMagAxisX + dezMinMagAxisY + dezMinMagAxisZ)/3

#########################################

###Eixo X do giroscopio a cada 2 minutos###

GyrAxisX = dataGyrAxisX.abs().groupby(dataGyrAxisX.index // step).mean()
doisMinGyrAxisX = GyrAxisX.loc[0,2]
quatroMinGyrAxisX = GyrAxisX.loc[1,2]
seisMinGyrAxisX = GyrAxisX.loc[2,2]
oitoMinGyrAxisX = GyrAxisX.loc[3,2]
dezMinGyrAxisX = GyrAxisX.loc[4,2]



###Eixo Y do giroscopio a cada 2 minutos###

GyrAxisY = dataGyrAxisY.abs().groupby(dataGyrAxisY.index // step).mean()
doisMinGyrAxisY = GyrAxisY.loc[0,3]
quatroMinGyrAxisY = GyrAxisY.loc[1,3]
seisMinGyrAxisY = GyrAxisY.loc[2,3]
oitoMinGyrAxisY = GyrAxisY.loc[3,3]
dezMinGyrAxisY = GyrAxisY.loc[4,3]

###Eixo Z do giroscopio a cada 2 minutos###

GyrAxisZ = dataGyrAxisZ.abs().groupby(dataGyrAxisZ.index // step).mean()
doisMinGyrAxisZ = GyrAxisZ.loc[0,4]
quatroMinGyrAxisZ = GyrAxisZ.loc[1,4]
seisMinGyrAxisZ = GyrAxisZ.loc[2,4]
oitoMinGyrAxisZ = GyrAxisZ.loc[3,4]
dezMinGyrAxisZ = GyrAxisZ.loc[4,4]

doisMinGyrAxisAvg = (doisMinGyrAxisX + doisMinGyrAxisY + doisMinGyrAxisZ)/3
quatroMinGyrAxisAvg = (quatroMinGyrAxisX + quatroMinGyrAxisY + quatroMinGyrAxisZ)/3
seisMinGyrAxisAvg = (seisMinGyrAxisX + seisMinGyrAxisY + seisMinGyrAxisZ)/3
oitoMinGyrAxisAvg = (oitoMinGyrAxisX + oitoMinGyrAxisY + oitoMinGyrAxisZ)/3
dezMinGyrAxisAvg = (dezMinGyrAxisX + dezMinGyrAxisY + dezMinGyrAxisZ)/3

###Calculando riscos sensores###

##2 minuto
if doisMinAccAxisAvg > 10 or doisMinMagAxisAvg >10 or doisMinGyrAxisAvg > 10:
  if doisMinAccAxisAvg > 30 or doisMinMagAxisAvg >30 or doisMinGyrAxisAvg > 30:
    riscoDoisMin = risco + 2
    print("trecho de 2 minutos: muito cuidado!")
  else:
    riscoDoisMin = risco + 1
    print("trecho de 2 minutos: cuidado")
  
##4 minuto  
if quatroMinAccAxisAvg > 10 or quatroMinMagAxisAvg >10 or quatroMinGyrAxisAvg > 10:
  if quatroMinAccAxisAvg > 30 or quatroMinMagAxisAvg >30 or quatroMinGyrAxisAvg > 30:
     riscoQuatroMin = risco + 2
     print("trecho de 4 minutos: muito cuidado!")
  else:
    riscoQuatroMin = risco + 1
    print("trecho de 4 minutos: cuidado")
 
##6 minuto
if seisMinAccAxisAvg > 10 or seisMinMagAxisAvg >10 or seisMinGyrAxisAvg > 10:
  if seisMinAccAxisAvg > 30 or seisMinMagAxisAvg >30 or seisMinGyrAxisAvg > 30:
    riscoSeisMin = risco + 2
    print("trecho de 6 minutos: muito cuidado!")
  else:
     riscoSeisMin = risco + 1
     print("trecho de 6 minutos: cuidado")

##8 minuto
if oitoMinAccAxisAvg > 10 or oitoMinMagAxisAvg >10 or oitoMinGyrAxisAvg > 10:
  if oitoMinAccAxisAvg > 30 or oitoMinMagAxisAvg >30 or oitoMinGyrAxisAvg > 30:
    riscoOitoMin = risco + 2
    print("trecho de 8 minutos: muito cuidado!")
  else:
    riscoOitoMin = risco + 1
    print("trecho de 8 minutos: cuidado")


if dezMinAccAxisAvg > 10 or dezMinMagAxisAvg >10 or dezMinGyrAxisAvg > 10:
  if dezMinAccAxisAvg > 30 or dezMinMagAxisAvg >30 or dezMinGyrAxisAvg > 30:
    riscoDezMin = risco + 2
    print("trecho de 10 minutos: muito cuidado!")
  else:
    riscoDezMin = risco + 1
    print("trecho de 10 minutos: cuidado")
  
#############################################


###Mostra Dados###

print("Velocidade média da rota:", velocidade,"m/s")
print("Metros Percorridos em 2 minutos:", distanciaMetrosDoisMinutos,"metros")
print("\n")
print("Dados a cada 2 minutos de coleta")
print("\n")
###2 Minutos###
print("2 minutos:")
print("\n")
print("Acelerometro")
print("Eixo X:" + str(doisMinAccAxisX))
print("Eixo Y:" + str(doisMinAccAxisY))
print("Eixo Z:" + str(doisMinAccAxisZ))
print("Média: " + str(doisMinAccAxisAvg))
print("\n")
print("Magnetometro")
print("Eixo X:" + str(doisMinMagAxisX))
print("Eixo Y:" + str(doisMinMagAxisY))
print("Eixo Z:" + str(doisMinMagAxisZ))
print("Média: " + str(doisMinMagAxisAvg))
print("\n")
print("Giro")
print("Eixo X:" + str(doisMinGyrAxisX))
print("Eixo Y:" + str(doisMinGyrAxisY))
print("Eixo Z:" + str(doisMinGyrAxisZ))
print("Média: " + str(doisMinGyrAxisAvg))
print("\n")
###4 Minutos###
print("4 minutos:")
print("\n")
print("Acelerometro")
print("Eixo X:" + str(quatroMinAccAxisX))
print("Eixo Y:" + str(quatroMinAccAxisY))
print("Eixo Z:" + str(quatroMinAccAxisZ))
print("Média: " + str(quatroMinAccAxisAvg))
print("\n")
print("Magnetometro")
print("Eixo X:" + str(quatroMinMagAxisX))
print("Eixo Y:" + str(quatroMinMagAxisY))
print("Eixo Z:" + str(quatroMinMagAxisZ))
print("Média: " + str(quatroMinMagAxisAvg))
print("\n")
print("Giro")
print("Eixo X:" + str(quatroMinGyrAxisX))
print("Eixo Y:" + str(quatroMinGyrAxisY))
print("Eixo Z:" + str(quatroMinGyrAxisZ))
print("Média: " + str(quatroMinGyrAxisAvg))
print("\n")
###6 minutos###
print("6 minutos:")
print("\n")
print("Acelerometro")
print("Eixo X:" + str(seisMinAccAxisX))
print("Eixo Y:" + str(seisMinAccAxisY))
print("Eixo Z:" + str(seisMinAccAxisZ))
print("Média: " + str(seisMinAccAxisAvg))
print("\n")
print("Magnetometro")
print("Eixo X:" + str(seisMinMagAxisX))
print("Eixo Y:" + str(seisMinMagAxisY))
print("Eixo Z:" + str(seisMinMagAxisZ))
print("Média: " + str(seisMinMagAxisAvg))
print("\n")
print("Giro")
print("Eixo X:" + str(seisMinGyrAxisX))
print("Eixo Y:" + str(seisMinGyrAxisY))
print("Eixo Z:" + str(seisMinGyrAxisZ))
print("Média: " + str(seisMinGyrAxisAvg))
print("\n")
###8 minutos###
print("8 minutos:")
print("\n")
print("Acelerometro")
print("Eixo X:" + str(oitoMinAccAxisX))
print("Eixo Y:" + str(oitoMinAccAxisY))
print("Eixo Z:" + str(oitoMinAccAxisZ))
print("Média: " + str(oitoMinAccAxisAvg))
print("\n")
print("Magnetometro")
print("Eixo X:" + str(oitoMinMagAxisX))
print("Eixo Y:" + str(oitoMinMagAxisY))
print("Eixo Z:" + str(oitoMinMagAxisZ))
print("Média: " + str(oitoMinMagAxisAvg))
print("\n")
print("Giro")
print("Eixo X:" + str(oitoMinGyrAxisX))
print("Eixo Y:" + str(oitoMinGyrAxisY))
print("Eixo Z:" + str(oitoMinGyrAxisZ))
print("Média: " + str(oitoMinGyrAxisAvg))
print("\n")
###10 minutos###
print("10 minutos:")
print("\n")
print("Acelerometro")
print("Eixo X:" + str(dezMinAccAxisX))
print("Eixo Y:" + str(dezMinAccAxisY))
print("Eixo Z:" + str(dezMinAccAxisZ))
print("Média: " + str(dezMinAccAxisAvg))
print("\n")
print("Magnetometro")
print("Eixo X:" + str(dezMinMagAxisX))
print("Eixo Y:" + str(dezMinMagAxisY))
print("Eixo Z:" + str(dezMinMagAxisZ))
print("Média: " + str(dezMinMagAxisAvg))
print("\n")
print("Giro")
print("Eixo X:" + str(dezMinGyrAxisX))
print("Eixo Y:" + str(dezMinGyrAxisY))
print("Eixo Z:" + str(dezMinGyrAxisZ))
print("Média: " + str(dezMinGyrAxisAvg))
print("\n")

print("pontos riscos:" + str(risco))
print("Riscos 2 minutos:" + str(riscoDoisMin))
print("Riscos 4 minutos:" + str(riscoQuatroMin))
print("Riscos 6 minutos:" + str(riscoSeisMin))
print("Riscos 8 minutos:" + str(riscoOitoMin))
print("Riscos 10 minutos:" + str(riscoDezMin))


header = ['Minutos', 'pontos']
result2min = ['2 Minutos', riscoDoisMin]
result4min = ['4 Minutos', riscoQuatroMin]
result6min = ['6 Minutos', riscoSeisMin]
result8min = ['8 Minutos', riscoOitoMin]
result10min = ['10 Minutos', riscoDezMin]


with open('result.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    writer.writerow(header)


    writer.writerow(result2min)
    writer.writerow(result4min)
    writer.writerow(result6min)
    writer.writerow(result8min)
    writer.writerow(result10min)
```
